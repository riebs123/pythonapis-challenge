```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy
from datetime import date 

date = date.today()

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1200)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1200)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))

cites_split = np.array_split(cities, 10)

```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = 'imperial'
query_url = url + "appid=" + weather_api_key + "&units=" + units + "&q="


city_data = []
city_number = 0


for city in cites_split[0]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')

time.sleep(60)

for city in cites_split[1]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')
        
time.sleep(60)

for city in cites_split[2]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')

time.sleep(60)

for city in cites_split[3]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')

time.sleep(60)

for city in cites_split[4]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')


time.sleep(60)

for city in cites_split[5]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')
        
time.sleep(60)

for city in cites_split[6]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')

time.sleep(60)

for city in cites_split[7]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')
        
time.sleep(60)

for city in cites_split[8]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')
        
        
time.sleep(60)

for city in cites_split[9]:
    try:
        response = requests.get(query_url + city).json()
        city = response['name']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        country = response['sys']['country']
        date = response['dt']
        city_number += 1 

        print('City: '+ city + ', City Number: ' + str(city_number))

        city_data.append([city, lat, lng, max_temp,humidity,
                         cloudiness, wind_speed, country, date])
        
        
        

    except:
        print('City not available')
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
northern_hemisphere_df = clean_city_df.loc[clean_city_df['Lat'] > 0]

x_values = northern_hemisphere_df['Lat']
y_values = northern_hemisphere_df['Max Temp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))

plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, "r-")

#plt.annotate(line_eq,(6,10), fontsize=15, color='r')
plt.xlabel('Latitude')
plt.ylabel('Max Temp')

print('-----------------------')
print('The linear regression equation is: '+ line_eq)
print('-----------------------')
print("The r-squared is: "+str(round(rvalue**2,2)))
print('-----------------------')


plt.show()
```

Graph Analysis: According to our linear regression equation, as latitude increases, max temp decreases. Based on the r-squared, the relationship is strong between these two measure compared to others simple regression equations we will see below.

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
southern_hemisphere_df = clean_city_df.loc[clean_city_df['Lat'] < 0]

x_values = southern_hemisphere_df['Lat']
y_values = southern_hemisphere_df['Max Temp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))

plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, "r-")

#plt.annotate(line_eq,(6,10), fontsize=15, color='r')
plt.xlabel('Latitude')
plt.ylabel('Max Temp')

print('-----------------------')
print('The linear regression equation is: '+ line_eq)
print('-----------------------')
print("The r-squared is: "+str(round(rvalue**2,2)))
print('-----------------------')


plt.show()
```

Graph Analysis: According to our linear regression equation, as latitude increases, max temp increases. According to the r-squared, our regression doesn't do a great job of accounting for the variance in data for southern hemisphere cities as it does for northern hemisphere cities.
